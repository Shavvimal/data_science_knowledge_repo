```python
from pathlib import Path
from PIL import Image
import pandas as pd
import numpy as np
import os


arr_of_heights_drive_1= []
arr_of_widths_drive_1= []

arr_of_heights_drive_2= []
arr_of_widths_drive_2= []

pathlist = Path('./Data Drive 1/').rglob('*.tif')
for path in pathlist:
     path_in_str = str(path)
     im = Image.open('./' + path_in_str.replace('\\', "/"))
     width, height = im.size
     arr_of_heights_drive_1 += [height]
     arr_of_widths_drive_1 += [width]

pathlist = Path('./Data Drive 2/').rglob('*.tif')
for path in pathlist:
     path_in_str = str(path)
     im = Image.open('./' + path_in_str.replace('\\', "/"))
     width, height = im.size
     arr_of_heights_drive_2 += [height]
     arr_of_widths_drive_2 += [width]
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
figure(figsize=(15, 10), dpi=80)
axes = plt.axes()
axes.grid()
plt.xlabel('width')
plt.ylabel('height')
plt.scatter(arr_of_widths_drive_2, arr_of_heights_drive_2, marker="x")
plt.scatter(arr_of_widths_drive_1, arr_of_heights_drive_1, marker="+")
plt.legend(['drive 2', 'drive 1'], loc=4)

plt.show()
```
