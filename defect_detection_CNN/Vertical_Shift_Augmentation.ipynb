```python
import os
from PIL import Image
import os.path, sys
import os
import random 
```

```python
file = random.choice(os.listdir('./drive/MyDrive/X-Ray/Bp_stetch_jpeg/'))
```

```python
# vertical shift image augmentation
from numpy import expand_dims
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import ImageDataGenerator
from matplotlib import pyplot

# load the image
img = load_img(f"./drive/MyDrive/X-Ray/Bp_stetch_jpeg/{file}")
# img = load_img(f"./dog.jpg")

# convert to numpy array
data = img_to_array(img)

# expand dimension to one sample
samples = expand_dims(data, 0)

# create image data augmentation generator
datagen = ImageDataGenerator(height_shift_range=0.5, fill_mode='wrap')

# prepare iterator
it = datagen.flow(samples, batch_size=1)

pyplot.figure(figsize=(9, 24), dpi=100)

# generate samples and plot
for i in range(9):
	# define subplot
	pyplot.subplot(330 + 1 + i)
	# generate batch of images
	batch = it.next()
	# convert to unsigned integers for viewing
	image = batch[0].astype('uint8')
	# plot raw pixel data
	pyplot.imshow(image)
 
# show the figure
pyplot.show()
```
